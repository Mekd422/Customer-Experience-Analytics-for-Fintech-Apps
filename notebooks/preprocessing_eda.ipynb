```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from config import DATA_PATHS


raw_df = pd.read_csv(DATA_PATHS['raw_reviews'])
clean_df = pd.read_csv(DATA_PATHS['processed_reviews'])
```

```python
# 3. Quick stats
print("Raw data shape:", raw_df.shape)
print("Cleaned data shape:", clean_df.shape)
print(clean_df['bank_name'].value_counts())
```

```python
# 4. Rating distribution
plt.figure(figsize=(8,5))
sns.countplot(data=clean_df, x='rating', hue='bank_name')
plt.title("Rating Distribution per Bank")
plt.show()
```

```python
# 5. Sample reviews
for bank in clean_df['bank_name'].unique():
    print(f"\nSample reviews from {bank}:")
    display(clean_df[clean_df['bank_name'] == bank].head(3))
```
